```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import csv
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
browser.get("https://www.tiket.com/kereta-api/cari?d=JAKARTA&dt=CITY&a=SGU&at=STATION&date=2023-06-22&adult=1&infant=0")
```

```python
soup = BeautifulSoup(browser.page_source,"lxml")
```

```python
data = soup.find('div', class_='train-list')
nama_kereta = data.find(class_='text-train-name').text
kelas_kereta = data.find(class_='text-train-class').text
stasiun_awal = data.find(class_='text-station').text
jam_awal = data.find(class_='text-time').text
jam_akhir = data.find(class_='text-time arrive').text
harga = data.find(class_='text-price').text
kursi = data.find(class_='seats-left').text
```

```python
datas = soup.find_all('div', class_='train-list')
```

```python
datas
```

```python
hasil = []
for data in datas:
    nama_kereta = data.find(class_='text-train-name').text
    kelas_kereta = data.find(class_='text-train-class').text
    stasiun_awal = data.find(class_='text-station').text
    jam_awal = data.find(class_='text-time').text
    jam_akhir = data.find(class_='text-time arrive').text
    harga = data.find(class_='text-price')
    if harga:
        harga = harga.text
        
    
    df = {
        'Nama': nama_kereta,
        'Kelas': kelas_kereta,
        'Keberangkatan': stasiun_awal,
        'JamAwal': jam_awal,
        'JamAkhir': jam_akhir,
        'Informasi': harga,
    }
    hasil.append(df)
```

```python
df = pd.DataFrame(hasil)
```
